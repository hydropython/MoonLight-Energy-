** Step 2: Data quality check:: Look for missing values, outliers, or incorrect entries (e.g., negative values where only positive should exist), especially in columns like GHI, DNI, and DHI and check for outliers, especially in sensor readings (ModA, ModB) and wind speed data (WS, WSgust).

```python
# Missing Value Checking
print(df.isnull().sum())
```

```python
# Outlier checking
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
outliers = find_outliers_IQR(df['GHI'])
outliers_DNI = find_outliers_IQR(df['DNI'])
outliers_DHI = find_outliers_IQR(df['DHI'])

print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))

print('number of outliers: '+ str(len(outliers_DNI)))

print('max outlier value: '+ str(outliers_DNI.max()))

print('min outlier value: '+ str(outliers_DNI.min()))

print('number of outliers: '+ str(len(outliers_DHI)))

print('max outlier value: '+ str(outliers_DHI.max()))

print('min outlier value: '+ str(outliers_DHI.min()))

```

```python
## Replace with interpolation POSTIVE OUTLIER NUMBERS
df.loc[outliers, 'GHI'] = None
df['GHI'] = df['GHI'].interpolate()
#df.loc[outliers_DNI, 'DNI'] = None
df['DNI'] = df['DNI'].interpolate()
#df.loc[outliers_DHI, 'DHI'] = None
df['DHI'] = df['DHI'].interpolate()
#df.to_csv('D:/10 ACADAMY KIFIYA/data (1)/data/df_data_cleaned.csv', index=False)

## set zero for the negative values
df['GHI'][df['GHI'] <= 0] = 0
df['DHI'][df['DHI'] <= 0] = 0
df['DNI'][df['DNI'] <= 0] = 0
df
```

```python
## Time Series Analysis: 
# Plot line graphs or area plots of GHI, DNI, DHI, and Tamb over time to observe patterns by month, trends throughout day, or anomalies, such as peaks in solar irradiance or temperature fluctuations. 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style='whitegrid')

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# Plot for GHI_1
axs[0].plot(df['Timestamp'], df['GHI'], color='tab:blue', linewidth=2)
axs[0].set_title('GHI', fontsize=14, fontweight='bold')
axs[0].set_ylabel('GHI (W/m²)', fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.6)

# Plot for GHI_2
axs[1].plot(df['Timestamp'], df['DHI'], color='tab:orange', linewidth=2)
axs[1].set_title('DHI2', fontsize=14, fontweight='bold')
axs[1].set_ylabel('GHI (W/m²)', fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.6)

# Plot for GHI_3
axs[2].plot(df['Timestamp'], df['DNI'], color='tab:green', linewidth=2)
axs[2].set_title('DNI', fontsize=14, fontweight='bold')
axs[2].set_xlabel('Date', fontsize=12)
axs[2].set_ylabel('GHI (W/m²)', fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.6)

# Rotate x-ticks for better readability
plt.xticks(rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
